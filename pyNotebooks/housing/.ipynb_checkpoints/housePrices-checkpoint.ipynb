# Boston House Prices

I'm going to build a model that summarizes the relationship between many variables and the price of a house. `sklearn` has some built in datasets and the boston house price dataset is one of them. Each record is a neighborhood and I'm going to focus on the following columns:

| Column Name | Description                                            |
|-------------|--------------------------------------------------------|
| CRIM        | per capita crime rate by town                          |
| price       | Median value of owner-occupied homes in \\$1,000        |
| TAX         | full-value property-tax rate per $10,000               |
| AGE         | proportion of owner-occupied units built prior to 1940 |
| RM          | average number of rooms per dwelling                   |
| PTRATIO     | pupil-teacher ratio by town                            |

Now I'll load the data.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from housing import load, investigate, featureCreate, model

df = load.importDat(['CRIM', 'price', 'TAX', 'AGE', 'RM', 'PTRATIO'])
df.describe()
```

This dataset has 506 rows and the above table contains some descriptive statistics.

- Average crime rate is around 3.6 crimes per person and the maximum was 88.97
- Median price of a house is around 2253, and the maximum observed price was 50k. 
- Average amount of tax per 10000 in value is 408
- Around 68% of houses were built prior to 1940, and there are some neighborhoods that are entirely made up of houses built prior to 1940.
- Most houses have around 6 rooms, and there isn't a lot of variation in that.
- The average pupil-teacher ratio is 18, which seems fairly low in my view.

No I'll do a quick check for missing values...

```python
df.isna().sum()
```

Looks good! Now lets take a look at the distributions of the pairs of variables.

```python
sns.pairplot(df[['TAX', 'CRIM', 'price', 'AGE', 'RM', 'PTRATIO']], corner=True);
```

A couple notes

How come the scatter plot only shows 2 points above 600 for `TAX`? After some further investigating, it appears that those 2 points are actually a lot more, just stacked on top of each other. I'll count the distinct combination of `TAX` and `PTRATIO` below.

```python
investigate.taxPratio(df)
```

Here we can see that those 2 points are actually 137 points, just stacked on top of each other. Let's keep that in mind as we move forward. Now we'll look at the correlation between the variables.

```python
sns.heatmap(df.corr());
```

It appears that `price` and `PTRATIO` are negatively correlated. This makes sense, as one would expect more expensive neighborhoods to have smaller classrooms (which some people think means the education is better but I won't draw that conclusion here).

Another thing to note is that `RM` and `price` are positively correlated. This also makes sense because I would expect larger houses to have higher prices.

## Modeling

No relationships in the scatter matrix suggested that there are any non-linear patterns, so I'll proceed by using a ordinary least squares model with no squared terms.

```python
olsModel = model.slr(featureCreate.normalize(df),
                      xVars=[x for x in df.columns if x != 'price'],
                      yVar='price')
print(olsModel.summary())

sns.barplot(x = 'var', y='coef', data=model.getCoefs(olsModel));
plt.title('Standardized Coefficient Estimates');
```

```python
model.getVIF(df[['CRIM', 'TAX', 'AGE', 'RM', 'PTRATIO']])
```

```python
model.getPlot(olsModel)
```

```python
model.getDFB(olsModel).query('hat_diag > .10')
```
